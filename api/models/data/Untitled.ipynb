```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

```python
anum_df = pd.read_csv('anum.csv')
```

```python
anum_df['hasil'] = anum_df['hasil'].replace("hati hati", 1)
anum_df['hasil'] = anum_df['hasil'].replace("lulus", 2)
anum_df['hasil'] = anum_df['hasil'].replace("tidak lulus", 0)
```

```python
anum_df
```

```python
anum_df.groupby('hasil').count()
```

```python
X = anum_df[['pras1', 'pras2']]
y = anum_df['hasil']
sm = SMOTE()
X_res, y_res = sm.fit_sample(X, y)
X_res = np.round(X_res)
y_res = np.round(y_res)
synthetic_df = pd.DataFrame(X_res)
synthetic_df.columns = X.columns
synthetic_df['y'] = y_res
```

```python
synthetic_df.groupby('y').count()
```

```python
synthetic_df['y'] = synthetic_df['y'].replace(1, "hati hati")
synthetic_df['y'] = synthetic_df['y'].replace(2, "lulus")
synthetic_df['y'] = synthetic_df['y'].replace(0, "tidak lulus")
```

```python
synthetic_df.groupby('y').count()
```

```python
synthetic_df.to_csv('anum.csv', index=False)
```
