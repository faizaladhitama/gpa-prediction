```python
from nb_model import NbModel
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("mk_prasyarat.csv")
data['npm'] = data['npm'].astype(str)
data.loc[data['npm'] == '1006826515']
```

```python
data
```

```python
def convert_kode_to_nilai(colname, kodename):
    data[colname] = np.nan
    for idx in data[kodename].dropna().index:
        lookup = data.loc[data['kd_mk'] == data[kodename][idx]]
        lookup = lookup.loc[lookup['npm'] == data['npm'][idx]]
        if len(lookup != 0):
            data.loc[idx,colname] = lookup['nilai_angka'].values[0]
```

```python
new_col = ['pras1', 'pras2', 'pras3', 'pras4']
count = 1
for pras in new_col:
    convert_kode_to_nilai(pras, 'kode_pras' + str(count))
    count = count + 1
data
```

```python
final = data.loc[:,['pras1', 'pras2', 'pras3', 'pras4', 'y']]
final = final.dropna(subset=['pras1', 'pras2', 'pras3', 'pras4'], how='all')
final['mean_pras'] = final.loc[:,['pras1', 'pras2', 'pras3', 'pras4']].mean(axis=1)
final.to_csv('final.csv', index=False)
```
