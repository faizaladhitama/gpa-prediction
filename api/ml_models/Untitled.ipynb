```python
from nb_model import NbModel
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("mk_prasyarat.csv")
data = data.dropna(subset=['y'])
data[data['y'].isnull()]
```

```python
data
```

```python
def convert_name_to_nilai(colscore, colname):
    data[colscore] = np.nan
    for idx in data[colname].dropna().index:
        lookup = data.loc[data['Nama_Matkul'] == data[colname][idx]]
        lookup = lookup.loc[lookup['npm'] == data['npm'][idx]]
        if len(lookup != 0):
            data.loc[idx,colscore] = lookup['nilai_angka'].values[0]
```

```python
for count in range(1,5):
    convert_name_to_nilai('nilai_pras' + str(count), 'nama_pras' + str(count))
data
```

```python
final = data.loc[:,['Nama_Matkul', 'nilai_pras1', 'nilai_pras2', 'nilai_pras3', 'nilai_pras4', 'y']]
final = final.dropna(subset=['nilai_pras1', 'nilai_pras2', 'nilai_pras3', 'nilai_pras4'], how='all')
final['mean_pras'] = final.loc[:,['nilai_pras1', 'nilai_pras2', 'nilai_pras3', 'nilai_pras4']].mean(axis=1)
final
```

```python
lookup = final[final['nilai_pras1'].isna() == True]
final = final.drop(lookup[lookup['nilai_pras2'].notna() == True].index)
```

```python
lookup = final[final['nilai_pras1'].isna() == True]
final = final.drop(lookup[lookup['nilai_pras3'].notna() == True].index)
```

```python
final.to_csv('final.csv', index=False)
```

```python
data = pd.read_csv("nilai.csv", delimiter='\t')
pras = pd.read_csv("prasyarat_matkul1.csv")
nm = pd.read_csv('info_matkul_final.csv')
nm = nm.drop_duplicates(subset=['kd_pasangan','Nama_Matkul'])
```

```python
data = pd.merge(data, nm,left_on='kd_mk', right_on='kd_pasangan', how='left')
data = data.dropna(subset=['kd_pasangan'])
data = data.drop('kd_pasangan', axis=1)
```

```python
pras = pd.merge(pras, nm,left_on='kd_mk', right_on='kd_pasangan', how='left')
pras = pd.merge(pras, nm,left_on='kode_pras1', right_on='kd_pasangan', how='left', suffixes=['',2])
pras = pd.merge(pras, nm,left_on='kode_pras2', right_on='kd_pasangan', how='left', suffixes=['',3])
pras = pd.merge(pras, nm,left_on='kode_pras3', right_on='kd_pasangan', how='left', suffixes=['',4])
pras = pd.merge(pras, nm,left_on='kode_pras4', right_on='kd_pasangan', how='left', suffixes=['',5])
pras = pras.drop('kd_pasangan', axis=1)
for suff in range(2,6):
    pras = pras.drop('kd_pasangan' + str(suff), axis=1)
    
```
