```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

```python
anum_df = pd.read_csv('anum.csv')
```

```python
anum_df
```

```python
anum_df['y'] = anum_df['y'].replace("hati hati", 1)
anum_df['y'] = anum_df['y'].replace("lulus", 2)
anum_df['y'] = anum_df['y'].replace("tidak lulus", 0)
```

```python
anum_df
```

```python
anum_df.groupby('y').count()
```

```python
X = anum_df[['pras1', 'pras2']]
y = anum_df['hasil']
sm = SMOTE()
X_res, y_res = sm.fit_sample(X, y)
X_res = np.round(X_res)
y_res = np.round(y_res)
synthetic_df = pd.DataFrame(X_res)
synthetic_df.columns = X.columns
synthetic_df['y'] = y_res
```

```python
synthetic_df.groupby('y').count()
```

```python
synthetic_df['y'] = synthetic_df['y'].replace(1, "hati hati")
synthetic_df['y'] = synthetic_df['y'].replace(2, "lulus")
synthetic_df['y'] = synthetic_df['y'].replace(0, "tidak lulus")
```

```python
synthetic_df.groupby('y').count()
```

```python
synthetic_df.to_csv('anum.csv', index=False)
```

```python
import os.path
import pickle
import pandas as pd
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
features = anum_df.loc[:, :'pras2']
target = anum_df.loc[:, 'y']
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.33, random_state=10)

clf.fit(features_train, target_train)
target_pred = clf.predict(features_test)
accuracy = accuracy_score(target_test, target_pred, normalize=True)
```
